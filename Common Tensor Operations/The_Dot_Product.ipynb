1. buat vector x dan y dengan ukuran 1x3

```python
import numpy as np
```

```python
# buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x = np.array([1, 2, 3])
x
```

```python
y = np.array([4, 5, 6])
y
```

```python
import torch
```

```python
# buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt = torch.tensor([3, 4, 5])
x_pt
```

```python
y_pt = torch.tensor([6, 7, 8])
y_pt
```

```python
import tensorflow as tf
```

```python
# buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf = tf.Variable([7, 8, 9])
x_tf
```

```python
y_tf = tf.Variable([10, 11, 12])
y_tf
```

2. hitung perkalian dot product antara vector x dan vector y

```python
# menghitung perkalian dot product menggunakan numpy
np.dot(x,y)
```

```python
# menghitung perkalian dot product menggunakan pytorch
torch.dot(x_pt, y_pt)
```

```python
# menghitung perkalian dot product menggunakan tensorflow
tf.reduce_sum(tf.multiply(x_tf, y_tf))
```
