### 1. Membuat matriks X ukuran 2x3, lalu jumlahkan semua elemen matrix menggunakan fungsi sum

1.1 membuat matriks X menggunakan numpy

```python
import numpy as np

# Membuat matriks 2x3
X = np.array([[1, 2, 3],
              [4, 5, 6]])

print(X)
print(f"Ukuran matriks: {X.shape}")
```

1.2 menjumlahkan semua elemen matriks menggunakan fungsi sum

```python
import numpy as np

X = np.array([[1, 2, 3],
              [4, 5, 6]])

# Menjumlahkan SEMUA elemen dalam matriks
total_sum = np.sum(X)

print(f"Matriks X")
print(f"Hasil penjumlahan semua elemen: {total_sum}")
```

1.3 membuat matriks X ukuran 2x3 dengan menggunakan tensorflow

```python
import tensorflow as tf

# Membuat matriks 2x3
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])

print("Matriks X_tf:")
print(X_tf)
print(f"Ukuran (Shape): {X_tf.shape}")
```

### 2. Hitung hasil penjumlahan seluruh elemen matriks X

```python
# hasil penjumlahan seluruh elemen matriks X
import numpy as np

X = np.array([[1, 2, 3],
              [4, 5, 6]])

# Menjumlahkan SEMUA elemen dalam matriks
total_sum = np.sum(X)

print(f"Matriks X")
print(f"Hasil penjumlahan semua elemen: {total_sum}")
```

```python
# hasil penjumlahan seluruh elemen matriks X_pt
import torch

# Membuat matriks 2x3 di PyTorch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

# Menghitung hasil penjumlahan seluruh elemen
total_sum_pt = torch.sum(X_pt)

print(f"Matriks X_pt:\n{X_pt}")
print(f"Hasil penjumlahan seluruh elemen (PyTorch): {total_sum_pt.item()}")
```

```python
# hasil penjumlahan seluruh elemen matriks X_tf
import tensorflow as tf

X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])

# Menghitung total penjumlahan
total_sum_tf = tf.reduce_sum(X_tf)

print(f"Hasil penjumlahan seluruh elemen (TF): {total_sum_tf.numpy()}")
```

### 3. Hitung hasil penjumlahan matrix X berdasarkan row nya

```python
# matrix X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

row_sum_np = np.sum(X, axis=1)
print(f"Hasil per baris (NumPy): {row_sum_np}")
```

```python
# matriks X_pt
import torch

# Membuat matriks 2x3
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

# Menghitung hasil penjumlahan berdasarkan baris (row)
row_sum = torch.sum(X_pt, dim=1)

print(f"Matriks X_pt:\n{X_pt}")
print(f"Hasil penjumlahan per baris: {row_sum}")
```

```python
# matrix X_tf
import tensorflow as tf

# matriks X_tf (2x3)
X_tf = tf.Variable([[1, 2, 3],
                    [4, 5, 6]])

# Menghitung penjumlahan berdasarkan baris
row_sum_tf = tf.reduce_sum(X_tf, axis=1)

print("Matriks X_tf:")
print(X_tf.numpy())
print(f"\nHasil penjumlahan per baris: {row_sum_tf.numpy()}")
# Output: [ 6 15]
```

### 4. hitung hasil penjumlahan matrix X berdasarkan kolomnya

```python
# matrix X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

col_sum_np = np.sum(X, axis=0)
print(f"Hasil per kolom (NumPy): {col_sum_np}")
```

```python
# matrix X_pt
import torch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

col_sum_pt = torch.sum(X_pt, dim=0)
print(f"Hasil per kolom (PyTorch): {col_sum_pt}")
```

```python
# matrix X_tf
import tensorflow as tf
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])

col_sum_tf = tf.reduce_sum(X_tf, axis=0)
print(f"Hasil per kolom (TF): {col_sum_tf.numpy()}")
```

### 5. Hitung hasil nilai yang dicari dari matrix X

5.1 nilai minimal

```python
# nilai minimal dari matrix X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

min_val = np.min(X)
# Atau: min_val = X.min()

print(f"Nilai minimal (NumPy): {min_val}")
```

```python
# nilai minimal dari matrix X_pt
import torch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

min_val_pt = torch.min(X_pt)

print(f"Nilai minimal (PyTorch): {min_val_pt.item()}")
```

```python
# nilai minimal dari matrix X_tf
import tensorflow as tf
X_tf = tf.Variable([[1, 2, 3],
                    [4, 5, 6]])

min_val_tf = tf.reduce_min(X_tf)

print(f"Nilai minimal (TF): {min_val_tf.numpy()}")
```

5.2 nilai maksimal

```python
# nilai maksimal dari matriks X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

max_val = np.max(X)
print(f"Nilai maksimal (NumPy): {max_val}")
```

```python
# nilai maksimal dari matrix X_pt
import torch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 8, 6]])

max_val_pt = torch.max(X_pt)
print(f"Nilai maksimal (PyTorch): {max_val_pt.item()}")
```

```python
# nilai maksimal dari matrix X_tf
import tensorflow as tf
X_tf = tf.Variable([[1, 9, 3],
                    [4, 5, 6]])

max_val_tf = tf.reduce_max(X_tf)
print(f"Nilai maksimal (TF): {max_val_tf.numpy()}")
```

5.3 nilai rata-rata

```python
# nilai rata - rata dari matrix X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

mean_val = np.mean(X)
print(f"Nilai rata-rata (NumPy): {mean_val}")
```

```python
# nilai rata - rata dari matrix X_pt
import torch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]], dtype = torch.float32)

mean_val_pt = torch.mean(X_pt)
print(f"Nilai rata-rata (PyTorch): {mean_val_pt}")
```

```python
# nilai rata - rata dari matrix X_tf
import tensorflow as tf
X_tf = tf.Variable([[1, 2, 3],
                    [4, 8, 6]])

mean_val_tf = tf.reduce_mean(X_tf)
print(f"Nilai rata-rata (TF): {mean_val_tf}")
```

### 6. Hitung hasil perkalian seluruh elemen matrix X

```python
# hasil perkalian seluruh elemen matrix X
import numpy as np
X = np.array([[1, 2, 3],
              [4, 5, 6]])

total_prod = np.prod(X)
print(f"Hasil perkalian (NumPy): {total_prod}")
```

```python
# hasil perkalian seluruh elemen matrix X_pt
import torch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

total_prod_pt = torch.prod(X_pt)
print(f"Hasil perkalian (PyTorch): {total_prod_pt}")
```

```python
# hasil perkalian seluruh elemen matrix X_tf
import tensorflow as tf
X_tf = tf.Variable([[1, 2, 3],
                    [4, 5, 6]])

total_prod_tf = tf.reduce_prod(X_tf)
print(f"Hasil perkalian (TF): {total_prod_tf}")
```
