### 1. buat matrix X dengan ukuran 2x3 lalu kalikan, jumlahkan, dan kurangkan dengan suatu skalar

1.1 Membuat matrix X dengan ukuran 2x3

```python
import numpy as np
```

```python
# buat matrix X dengan ukuran 2x3 dengan menggunakan numpy
X = np.array([[1, 2,3], [5, 6, 7]])
X
```

```python
import torch
```

```python
# buat matrix X_pt dengan ukuran 2x3 dengan menggunakan pytorch
X_pt = torch.tensor([[1, 4, 6], [5, 6, 7]])
X_pt
```

```python
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 dengan menggunakan tensorflow
X_tf = tf.Variable([[1, 3, 6], [8, 6, 7]])
X_tf
```

1.2 Mengalikan Matriks X dengan suatu skalar

```python
# kalikan matriks X dengan suatu skalar
X * 6 # 6 adalah nilai scalar
```

```python
# kalikan matriks X_pt dengan suatu skalar
torch.mul(X_pt, 4)
```

```python
# kalikan matriks X_tf dengan suatu skalar
tf.multiply(X_tf, 3)
```

1.3 Menjumlahan matrix x dengan suatu skalar

```python
# jumlahkan matriks X dengan suatu skalar
X + 2
```

```python
# jumlahkan matriks X dengan suatu pytorch
torch.add(X_pt, 5)
```

```python
# jumlahkan matriks X dengan suatu tensorflow
tf.add(X_tf, 7)
```

1.4 Mengurangkan matriks X dengan suatu matriks

```python
# kurangkan matriks X dengan suatu skalar
X - 1
```

```python
# kurangkan matriks X_pt dengan suatu skalar
torch.sub(X_pt, 4)
```

```python
# kurangkan matriks X_tf dengan suatu skalar
tf.subtract(X_tf, 8)
```

### 2. Buat matrix Y dengan ukuran 2x3 lalu kalikan dengan matrix X

2.1 Buat matriks Y ukuran 2x3

```python
# buat matrix Y dengan ukuran 2x3 dengan menggunakan numpy
Y = np.array([[1, 2, 3], [5, 6, 7]])
Y
```

```python
# buat matrix Y_pt dengan ukuran 2x3 dengan menggunakan pytorch
Y_pt = torch.tensor([[9, 6, 4], [4, 3, 7]])
Y_pt
```

```python
# buat matrix X_tf dengan ukuran 2x3 dengan menggunakan tensorflow
Y_tf = tf.Variable([[1, 3, 6], [8, 6, 7]])
Y_tf
```

2.2 Kalikan matrix X dengan Y

```python
# kalikan matriks X dan Y
X * Y
```

```python
# kalikan matriks X_pt dan Y_pt
torch.mul(X_pt, Y_pt)
```

```python
# kalikan matriks X_tf dan Y_tf
tf.multiply(X_tf, Y_tf)
```
