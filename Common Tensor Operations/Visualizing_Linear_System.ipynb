1. buat visualisasi dari persamaan berikut
y = 2x + 1 &
3x - y = 5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # mulai, selesai, n titik
```

```python
# persamaan pertama
y1 = (2 * x) + 1
```

```python
# persamaan kedua
y2 = (3 * x) - 5
```

```python
# grafik
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 15])
ax.set_ylim([0, 20])
ax.plot(x, y1, c='blue')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='red', linestyle='--')
_ = plt.axhline(y=13, color='red', linestyle='--')
```
