1. Buat matrix x dengan ukuran 2x3

```python
import numpy as np
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[2, 3, 4], [1, 2, 3]])
X
```

```python
import torch
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[2, 3, 4], [1, 2, 3]])
X_pt
```

```python
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[2, 3, 4], [1, 2, 3]])
X_tf
```
