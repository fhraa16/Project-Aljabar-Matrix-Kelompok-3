3. lakukan soal nomor 2 menggunakan pytorch dan tensorflow

```python
import torch
import tensorflow as tf
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_pt
x_pt = torch.tensor([[91, 19, 17]])
x_pt = x_pt.T
x_pt
```

```python
# tampilkan ukuran dari vector x_pt
x_pt.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_pt
y_pt = torch.tensor([[58],[61],[5]])
y_pt = y_pt.T
y_pt
```

```python
# tampilkan ukuran dari vector x_pt
x_pt.shape
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_tf
x_tf = tf.Variable([[51, 15, 11]])
x_tf = tf.transpose(x_tf)
x_tf
```

```python
# tampilkan ukuran dari vector x_tf
x_tf.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_tf
y_tf = tf.Variable([[50],[71],[55]])
y_pt = tf.transpose(y_tf)
y_pt
```

```python
# tampilkan ukuran dari vector x_tf
x_tf.shape
```
