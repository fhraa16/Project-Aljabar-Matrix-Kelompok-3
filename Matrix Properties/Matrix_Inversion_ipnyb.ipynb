### 1. Buat matrix X ukuran 2x2, lalu diinverskan, dan hasil invers tersebut dikalikan dengan matrix X

````python
import numpy as np
import matplotlib.pyplot as plt
````

1.1 buat matrikx X ukuran 2x2 menggunakan numpy

````python
# matrix X ukuran 2x2
X = np.array([[1, 2], [-3, -4]])
X
````

1.2 buat invers dari matrix X dan beri nama Xinv menggunakan numpy

````python
# buat invers dari matrix X
Xinv = np.linalg.inv(X)
Xinv
````

1.3 kalikan X dan Xinv menggunakan numpy

````python
# X dikali Xinv
np.dot(X, Xinv)
````

Itu bukan salah, tapi efek pembulatan (floating point), Itu tetap benar, cuma error pembulatan komputer.

### 2. Selesaikan persamaan berikut menggunakan numpy, tensorflow, dan pytorch

Dari gambar, sistem persamaannya adalah:
*   4b  + 2c =4
*   −5b − 3c =−7



> Kita ubah ke bentuk matriks:
```
|4 ,  2|  |b| = | 4|
|-5, -3|  |c| = |-7|

````python
# numpy
A = np.array([[4, 2],
              [-5, -3]])

B = np.array([4, -7])

solusi = np.linalg.solve(A, B)

b, c = solusi
print("b =", b)
print("c =", c)
````

````python
# tensorflow
import tensorflow as tf
A = tf.constant([[4.0, 2.0],
                 [-5.0, -3.0]])

B = tf.constant([[4.0],
                 [-7.0]])

solusi = tf.linalg.solve(A, B)

b = solusi[0][0].numpy()
c = solusi[1][0].numpy()

print("b =", b)
print("c =", c)
````

````python
# PyTorch
import torch
A = torch.tensor([[4.0, 2.0],
                  [-5.0, -3.0]])

B = torch.tensor([[4.0],
                  [-7.0]])

solusi = torch.linalg.solve(A, B)

b = solusi[0][0].item()
c = solusi[1][0].item()

print("b =", b)
print("c =", c)
````
