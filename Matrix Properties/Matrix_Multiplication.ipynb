### 1. buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1

numpy

```python
import numpy as np
```

```python
A = np.array([[4, 5, 6], [7, 8, 9]])
A
```

```python
b = np.array([[3], [1], [4]])
b
```

pytorch

```python
import torch
```

```python
A_pt = torch.tensor([[4, 5, 6], [7, 8, 9]])
A_pt
```

```python
b_pt = torch.tensor([[3], [1], [4]])
b_pt
```

tensorflow

```python
import tensorflow as tf
```

```python
A_tf = tf.Variable([[4, 5, 6], [7, 8, 9]])
A_tf
```

```python
b_tf = tf.Variable([[3], [1], [4]])
b_tf
```

### 2.  tampilkan ukuran matrix A dan vector b

```python
A.shape
```

```python
b.shape
```

### 3. kalikan matrix A dengan vector b

```python
# numpy
np.dot(A, b)
```

```python
# pytorch
torch.matmul(A_pt, b_pt)
```

```python
# tensorflow
tf.matmul(A_tf, b_tf)
```

### 4. buat matrix B dengan ukuran 3x3

```python
# numpy
B = np.array([[4, 5, 8], [3, 5, 9], [3, 2, 9]])
B
```

```python
# pytorch
B_pt = torch.tensor([[4, 5, 8], [3, 5, 9], [3, 2, 9]])
B_pt
```

```python
# tensorflow
B_tf = tf.Variable([[4, 5, 8], [3, 5, 9], [3, 2, 9]])
B_tf
```

### 5. kalikan matrix A dengan matrix B


```python
# numpy
np.dot(A, B)
```

```python
# pytorch
torch.matmul(A_pt, B_pt)
```

```python
# tensorflow
tf.matmul(A_tf, B_tf)
```
